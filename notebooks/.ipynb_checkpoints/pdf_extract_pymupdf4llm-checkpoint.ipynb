```python
root_dir = pathlib.Path.cwd().parent
pdf_file = root_dir / "data" / "raw" / "ch1_ch14_Brain_and_behavior.pdf"
output_mdfile = root_dir / "data" / "processed" / "ch1_ch14_Brain_and_behavior.md"
```

## simple version (without header and footer treatment)

```python
md_text = pymupdf4llm.to_markdown(pdf_file)
```

```python
pathlib.Path(output_mdfile).write_bytes(md_text.encode())
```

## header & footer treatment

```python
md_text_nohf = pymupdf4llm.to_markdown(pdf_file,header=False, footer=False)
```

```python
output_mdfile_nohf = root_dir / "data" / "processed" / "ch1_ch14_Brain_and_behavior_nohf.md"
```

```python
pathlib.Path(output_mdfile_nohf).write_bytes(md_text_nohf.encode())
```
